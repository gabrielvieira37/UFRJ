# F = [ x² , y ]

```python
def fxy(x,y):
    v = np.zeros(2)
    v[0] = x*x
    v[1] = y
    return v
# f(x,y) = (x²,x+y)
def jxy(x,y):
    v = np.zeros((2,2))
    v[0,0] = 2*x
    v[1,1] = 1
    return v
#j(x,y)= [dx²/dx = 2x,dx²/dy = 0][dy/dx = 0,dy/dy = 1]
```

```python
numInt = 5
np.random.seed(123)
pontos = []  
diff=1
x0 = np.random.rand(2)
while diff > 10**-4 : 
    print(x0)
    pontos.append(x0)
    Fx = fxy(x0[0],x0[1])
    Dij = jxy(x0[0],x0[1])
    y = np.linalg.solve(Dij,-Fx)
    x1 = y + x0
    diff = np.linalg.norm(x1-x0)
    x0 = np.copy(x1)
    
print(x1)
print('diff: ',diff)
```

# F = [ sin(x) , e^(-y) - 0.9y ]

```python
def fxy2(x,y):
    v = np.zeros(2)
    v[0] = np.sin(x)
    v[1] = np.exp(-y) -0.9*y
    return v
# f(x,y) = (sin(x)+3,e^y)
def jxy2(x,y):
    v = np.zeros((2,2))
    v[0,0] = np.cos(x)
    v[1,1] = -np.exp(-y) - 0.9
    return v
# J(x,y) = [d(sen(x))/dx = cos(x), d(sen(x))/dy = 0]
#          [d(e^-y)-0.1 /dx = 0,d(e^y)-0.1/dy = -e^-y]
```

```python
numInt = 15
diff = 1    
np.random.seed(1234)
pontos = []
x0 = np.random.randint(5, size=2)
#x0[1] = -x0[1]

while diff > 10**-4 : 
    print(x0)
    pontos.append(x0)
    Fx = fxy2(x0[0],x0[1])
    Dij = jxy2(x0[0],x0[1])
    y = np.linalg.solve(Dij,-Fx)
    x1 = y + x0
    diff = np.linalg.norm(x1-x0)
    x0 = np.copy(x1)
    
    

print(x1)
print('diff: ',diff)
```

```python
# converge para todos os valores positivos
# não converge para um zero de função com valores de x negativos
# converge para um zero de função mesmo se Y for negativo, mas X positivo
# quando feito para e^-y, a matriz Jacobiana acabava caindo em uma matriz singular.
```

# F = [sin(x*y) - x² , e^(-x*y) - y²]

```python
def fxy3(x,y):
    v = np.zeros(2)
    v[0] = np.sin(x*y)  - x**2
    v[1] = np.exp(-(x*y)) - y**2
    return v
# f(x,y) = (sin(x)+3,e^y)
def jxy3(x,y):
    v = np.zeros((2,2))
    v[0,0] = np.cos(x*y)*y - 2*x
    v[0,1] = np.cos(x*y)*x
    v[1,0] = -y*np.exp(-y)
    v[1,1] = -x*np.exp(-y) - 2*y
    return v
# J(x,y) = [d(sen(x))/dx = ycos(xy), d(sen(yx))/dy = xcos(xy)]
#          [d(e^-xy)-0.1 /dx = -ye^-xy,d(e^xy)-0.1/dy = -xe^-xy]
```

```python
# np.random.seed(1234)
np.random.seed(1)
pontos = []
x0 = np.random.randint(low=1,high =4, size=2) 
# x0 = np.random.rand(2)
# x0[1] = -x0[1]
diff =1
while diff > 10**-4 : 
    print(x0)
    pontos.append(x0)
    Fx = fxy3(x0[0],x0[1])
    Dij = jxy3(x0[0],x0[1])
    y = np.linalg.solve(Dij,-Fx)
    x1 = y + x0
    diff = np.linalg.norm(x1-x0)
    x0 = np.copy(x1)
print(x1)
print('diff: ',diff)
```
