# Testei 3 alfas diferentes para cada uso da regularização de Tikhonov - (0.1 , 1 , 100 ) , quanto maior o alfa a curva parecia mais suave e mais "generalizada", a mudança do alfa também pode aumentar as amplitude dos vales vide a matriz quadrada ( coloquei um alfa = 1000 a mais para explicitar essa caracteristica ). 

```python
import SVD 
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def pseudo_solve(A,b):
    ps = np.linalg.pinv(A)
    x = ps.dot(b)
    return x
    
                 
```

```python
def tikhonov(A,b,alpha):
    m = A.T.dot(A) 
    id = np.identity(m.shape[0])
    alphaI = alpha*id
    m = m+alphaI
    atB = A.T.dot(b)
    x = np.linalg.solve(m, atB)
    return x
    
    
```

```python
def f(t):
    f = np.zeros((t.shape[0],A.shape[1]))
    f[:,0] += 1
    for i in range(t.shape[0]):
        for j in range(1,f.shape[1]):
            f[i,j] = t[i] ** j
    f= f.dot(x)
    return f
```

```python
A = np.array(([1,2,4],[1,3,9]))
b = np.array(([4],[6]))
```

```python
x = tikhonov(A,b,0.0001)
x = pseudo_solve(A,b)


```

```python
axisX = A[:,1]
axisY = b
```

```python
t1 = np.arange(0.0, 6.0, 0.1)
```

```python
plt.plot(t1,f(t1),axisX,axisY,'bo')
plt.show()
```
